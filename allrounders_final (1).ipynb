```python
import pandas as pd
data=pd.read_csv("allrounders.txt")
print(f"\n {5*"-"} LIST OF TOP 20 ALL ROUNDERS {5*"-"} \n")
x=1    #variable for counting
for player in data["Player"]:
    print(x,player)
    x+=1
print(f"\n {5*"-"} HIGHEST WICKET TAKER {5*"-"} \n")
print(data.loc[[data["Wickets"].idxmax()],["Player","Wickets","Country"]])     #remember (row,coloumn)
print(f"\n {5*"-"} HIGHEST RUN GETTER {5*"-"} \n")
print(data.loc[[data["Runs"].idxmax()],["Player","Runs","Country"]])
print(f"\n {5*"-"} BEST BATTING AVG {5*"-"} \n")
print(data.loc[[data["Batting_Average"].idxmax()],["Player","Batting_Average","Country"]])
print(f"\n {5*"-"} BEST BOWLING AVG {5*"-"} \n")
print(data.loc[[data["Bowling_Average"].idxmin()],["Player","Bowling_Average","Country"]])
print(f"\n {5*"-"} BEST STRIKERATE {5*"-"} \n")
print(data.loc[[data["Strike_Rate"].idxmax()],["Player","Strike_Rate","Country"]])
print(f"\n {5*"-"} LEAST ECONOMY {5*"-"} \n")
print(data.loc[[data["Economy"].idxmin()],["Player","Economy","Country"]])

#finding best allrounder usingg score system
data["score"]=data["Runs"]*0.01+data["Wickets"]*0.35+data["Batting_Average"]*0.2-data["Bowling_Average"]*0.15-data["Economy"]*0.15+data["Strike_Rate"]*0.14
data["score"]=data["score"].round(2)
top5 = data.sort_values(by="score", ascending=False).head(5)   #sorting in desc order


print(f"\n {5*"-"} BEST 5 ALL ROUNDERS LIST {5*"-"} \n")
print("NAME , COUNRY ,SCORE")
for p,c,s in zip(top5["Player"],top5["Country"],top5["score"]):
    print(f"{p:20} {c:12} {s}")
    
```
